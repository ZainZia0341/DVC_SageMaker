```python
import os
import json
import boto3
from typing import Any, Dict

class SageMakerJumpStartLLM:
    def __init__(self, endpoint_name: str, region: str = "us-east-1"):
        """
        Initialize the SageMaker JumpStart LLM with the endpoint name and region.
        """
        self.endpoint_name = endpoint_name
        self.region = region
        self.client = boto3.client("sagemaker-runtime", region_name=region)

    def invoke(self, prompt: str) -> Dict[str, Any]:
        """
        Invoke the SageMaker endpoint with the provided prompt.
        
        Args:
            prompt (str): The prompt or message to send to the model.
            
        Returns:
            Dict: The JSON response from the endpoint.
        """
        # Prepare the payload. Adjust the key if your model expects a different format.
        payload = {"prompt": prompt}
        response = self.client.invoke_endpoint(
            EndpointName=self.endpoint_name,
            ContentType="application/json",
            Body=json.dumps(payload)
        )
        result = json.loads(response["Body"].read().decode())
        return result

# Optional: Create a global instance.
# Make sure your SAGEMAKER_ENDPOINT environment variable is set with your real endpoint name.
endpoint_name = os.getenv("SAGEMAKER_ENDPOINT")
if not endpoint_name:
    raise ValueError("Please set the SAGEMAKER_ENDPOINT environment variable.")

# Optionally, set the region via an environment variable or default it.
aws_region = os.getenv("AWS_REGION", "us-east-1")

# Create a global LLM instance that you can import elsewhere.
llm = SageMakerJumpStartLLM(endpoint_name=endpoint_name, region=aws_region)
```

```python
# src/inference.py

import os
import io
import json
import base64
from PIL import Image
import torch
from transformers import AutoProcessor, AutoModelForVisionSeq2SeqLM

def model_fn(model_dir):
    """
    Load the model & processor from the nested folder.
    model_dir will be '/opt/ml/model'
    """
    subfolder = os.path.join(model_dir, "Llama-3.2-11B-Vision-Instruct-bnb-4bit")
    processor = AutoProcessor.from_pretrained(subfolder)
    model = AutoModelForVisionSeq2SeqLM.from_pretrained(subfolder)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    return {"processor": processor, "model": model, "device": device}

def _decode_image(b64_string):
    """
    Convert a base64encoded PNG/JPG into a PIL image.
    """
    data = base64.b64decode(b64_string)
    return Image.open(io.BytesIO(data)).convert("RGB")

def input_fn(request_body, request_content_type):
    """
    Deserialize the JSON request into a Python dict.
    Expecting: {"image": "<base64-png>", "text": "Your prompt"}
    """
    if request_content_type == "application/json":
        return json.loads(request_body)
    raise ValueError(f"Unsupported content type: {request_content_type}")

def predict_fn(input_data, model_bundle):
    """
    Run a single forward pass: vision+text  generated tokens  decoded text.
    """
    # 1) Decode image
    img = _decode_image(input_data["image"])
    # 2) Get optional text prompt
    text = input_data.get("text", "")
    # 3) Tokenize & prepare tensors
    inputs = model_bundle["processor"](images=img, text=text, return_tensors="pt")
    # 4) Move to GPU/CPU
    inputs = {k: v.to(model_bundle["device"]) for k, v in inputs.items()}
    # 5) Generate
    generated_ids = model_bundle["model"].generate(**inputs)
    # 6) Decode
    return model_bundle["processor"].batch_decode(generated_ids, skip_special_tokens=True)

def output_fn(prediction, response_content_type):
    """
    Serialize the Python list of strings into JSON.
    """
    if response_content_type == "application/json":
        return json.dumps(prediction), "application/json"
    raise ValueError(f"Unsupported response content type: {response_content_type}")
```

```python
# install prerequisites if you haven't already:
# pip install transformers accelerate safetensors

import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline,
)

def main():
    # 1) point to your local model folder
    model_path = r"D:\zain\gemma-2-2b-bnb-4bit"

    # 2) load tokenizer & model
    tokenizer = AutoTokenizer.from_pretrained(model_path, use_fast=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_path,
        torch_dtype=torch.float16,     # use fp16 if your GPU supports it
        trust_remote_code=True,        # allow any repo-defined code
        device_map="auto",             # auto-place layers on GPUs/CPU
    )

    # 3a) quick test via  pipeline
    gen = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        device_map="auto",
        trust_remote_code=True,
    )

    prompt = "Once upon a time"
    out = gen(
        prompt,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.8,
        top_p=0.9,
    )
    print("=== pipeline output ===")
    print(out[0]["generated_text"])

    # 3b) or manual generate
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    generate_ids = model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.8,
        top_p=0.9,
    )
    text = tokenizer.batch_decode(generate_ids, skip_special_tokens=True)[0]
    print("\n=== manual generate ===")
    print(text)

if __name__ == "__main__":
    main()
```
